Solving the Problem Recursively

```python
import time
 
def fib_rec(f1, f2, n):
    if n == 0:
        return f1
    else:
        return fib_rec(f2, f1 + f2, n - 1)
 
def fib_v2(num):
    return fib_rec(0, 1, num)
n = 10
while n:
  t1 = time.time()
  fib_v2(n)
  t2 = time.time()
  print("this is ",n)
  print(fib_v2(n))
  print('{} seconds'.format((t2 - t1)))
  n = n+10
```
